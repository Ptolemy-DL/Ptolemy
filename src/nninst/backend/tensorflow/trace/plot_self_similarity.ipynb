```python
import numpy as np
from nninst.backend.tensorflow.trace.alexnet_imagenet_class_trace import alexnet_imagenet_class_trace_compact, \
    alexnet_imagenet_self_similarity
from nninst.backend.tensorflow.trace.alexnet_imagenet_inter_class_similarity import alexnet_imagenet_inter_class_similarity
from nninst.backend.tensorflow.trace.inception_v4_imagenet_class_trace import inception_v4_imagenet_self_similarity
from nninst.backend.tensorflow.trace.resnet_18_cifar10_inter_class_similarity import resnet_18_cifar10_inter_class_similarity
from nninst.backend.tensorflow.trace.vgg_16_imagenet_class_trace import vgg_16_imagenet_self_similarity
```

```python
trace_type = None
trace_parameter = None
threshold = 0.5
label = "import"
variant = None
total_time = 0
```

```python
similarity = alexnet_imagenet_self_similarity(
    threshold, label, variant=variant
).load()
similarity.shape
```

```python
%matplotlib inline
import seaborn as sns
similarity = similarity[::10, ::10]
ax = sns.heatmap(similarity, cmap="Greys", vmax=0.65, annot=True)
ax.set(xlabel='Class', ylabel='Class')
fig = ax.get_figure()
fig.show()
```

```python
np.savetxt("../../../plot/alexnet_imagenet_self_similarity.csv", similarity, delimiter=",")
```
